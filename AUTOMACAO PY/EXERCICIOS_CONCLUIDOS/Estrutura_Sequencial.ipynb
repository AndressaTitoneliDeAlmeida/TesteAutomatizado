#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
# Peça ao usuário a temperatura em graus Fahrenheit
fahrenheit = float(input("Digite a temperatura em graus Fahrenheit: "))

# Converta para graus Celsius usando a fórmula
celsius = (5/9) * (fahrenheit - 32)

# Exiba o resultado
print(f"A temperatura em graus Celsius é: {celsius:.2f}°C")
```

#### 10. Tendo como dados de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, usando a seguinte fórmula:

#### 11. Tendo como dado de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:
##### a. Para homens: $P = 72,7h - 58$
##### b. Para mulheres: $P = 62,1h - 44,7$

```python
# Solicite ao usuário a altura em metros
altura = float(input("Digite a altura em metros: "))

# Solicite ao usuário o sexo (M para masculino, F para feminino)
sexo = input("Digite o sexo M para masculino, F para feminino: ").upper()

# Verifique o sexo e calcule o peso ideal de acordo com a fórmula apropriada
if sexo == "M":
    peso_ideal = 72.7 * altura - 58
elif sexo == "F":
    peso_ideal = 62.1 * altura - 44.7
else:
    print("Sexo não reconhecido. Use 'M' para masculino ou 'F' para feminino.")
    peso_ideal = None

# Exiba o resultado, se o sexo for reconhecido
if peso_ideal is not None:
    print(f"Seu peso ideal é: {peso_ideal:.2f} kg")
```

#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
# Solicite ao usuário o valor do pagamento por hora
pagamento_hora = float(input("Digite o valor do pagamento por hora: "))

# Solicite ao usuário o número de horas trabalhadas no mês
horas_trabalhadas_mes = float(input("Digite o número de horas trabalhadas no mês: "))

# Calcular o salário mensal
salario_mensal = pagamento_hora * horas_trabalhadas_mes

# Exiba o resultado
print(f"Seu salário mensal é: R$ {salario_mensal:.2f}")
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
# Solicite ao usuário o valor do pagamento por hora
pagamento_hora = float(input("Digite o valor do pagamento por hora: "))

# Solicite ao usuário o número de horas trabalhadas no mês
horas_trabalhadas_mês = float(input("Digite o número de horas trabalhadas no mês: "))

# Calcule o salário bruto
salario_bruto = pagamento_hora * horas_trabalhadas_mês

# Exiba o resultado
print(f"Seu salário bruto é: R$ {salario_bruto:.2f}")
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
# Solicite ao usuário o valor do pagamento por hora
pagamento_hora = float(input("Digite o valor do pagamento por hora: "))

# Solicite ao usuário o número de horas trabalhadas no mês
horas_trabalhadas_mês = float(input("Digite o número de horas trabalhadas no mês: "))

# Calcule o salário bruto
salario_bruto = pagamento_hora * horas_trabalhadas_mês

# Calcule o desconto do Imposto de Renda (IR) de 11%
desconto_ir = 0.11 * salario_bruto

# Exiba o resultado
print(f"Seu desconto de Imposto de Renda (IR) é: R$ {desconto_ir:.2f}")
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
# Solicite ao usuário o valor do pagamento por hora
pagamento_hora = float(input("Digite o valor do pagamento por hora: "))

# Solicite ao usuário o número de horas trabalhadas no mês
horas_trabalhadas_mês = float(input("Digite o número de horas trabalhadas no mês: "))

# Calcule o salário bruto
salario_bruto = pagamento_hora * horas_trabalhadas_mês

# Calcule o desconto do INSS de 8%
desconto_inss = 0.08 * salario_bruto

# Exiba o resultado
print(f"Seu desconto de INSS é: R$ {desconto_inss:.2f}")
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
# Solicite ao usuário o valor do pagamento por hora
pagamento_hora = float(input("Digite o valor do pagamento por hora: "))

# Solicite ao usuário o número de horas trabalhadas no mês
horas_trabalhadas_mês = float(input("Digite o número de horas trabalhadas no mês: "))

# Calcule o salário bruto
salario_bruto = pagamento_hora * horas_trabalhadas_mês

# Calcule o desconto do Sindicato de 5%
desconto_sindicato = 0.05 * salario_bruto

# Exiba o resultado
print(f"Seu desconto do Sindicato é: R$ {desconto_sindicato:.2f}")
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
# Solicite ao usuário o valor da taxa de pagamento por hora
pagamento_hora = float(input("Digite o valor do pagamento por hora: "))

# Solicite ao usuário o número de horas trabalhadas no mês
horas_trabalhadas_mês = float(input("Digite o número de horas trabalhadas no mês: "))

# Calcule o salário bruto
salário_bruto = pagamento_hora * horas_trabalhadas_mês

# Calcule o desconto do Imposto de Renda (IR) de 11% do salário bruto
desconto_ir = 0.11 * salário_bruto

# Calcule o desconto do INSS (8% do salário bruto)
desconto_inss = 0.08 * salário_bruto

# Calcule o desconto do Sindicato de 5%
desconto_sindicato = 0.05 * salario_bruto

# Calcule o salário líquido (salário bruto - descontos)
salário_líquido = salário_bruto - (desconto_ir + desconto_inss + desconto_sindicato)

# Exiba os resultados
print(f"Salário Bruto: R$ {salário_bruto:.2f}")
print(f"Desconto do INSS: R$ {desconto_inss:.2f}")
print(f"Desconto do Imposto de Renda (IR): R$ {desconto_ir:.2f}")
print(f"Seu desconto do Sindicato é: R$ {desconto_sindicato:.2f}")
print(f"Salário Líquido: R$ {salário_líquido:.2f}")
```

#### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
# Solicite ao usuário o tamanho em metros quadrados da área a ser pintada
area_a_ser_pintada = float(input("Digite o tamanho da área a ser pintada em metros quadrados: "))

# Calcule a quantidade de tinta necessária (litros)
litros_de_tinta_necessarios = area_a_ser_pintada / 3

# Calcule a quantidade de latas de tinta necessárias
latas_de_tinta_necessarias = litros_de_tinta_necessarios / 18

# Calcule o preço total
preco_total = latas_de_tinta_necessarias * 80

# Exiba o resultado
print(f"Quantidade de latas de tinta a serem compradas: {latas_de_tinta_necessarias:.2f}")
print(f"Preço total: R$ {preco_total:.2f}")
```
